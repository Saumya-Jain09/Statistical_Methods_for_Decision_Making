```python
def detect_outlier(col):
  Q1,Q3=np.percentile(col,[25,75])
  IQR=Q3-Q1
  lower_range=Q1-(1.5*IQR)
  upper_range=Q3+(1.5*IQR)
  return lower_range,upper_range
```

```python
lr,ur=detect_outlier(data1['Total_salary'])
```

```python
print (ur,lr)
```

```python
data1['Total_salary']=np.where(data1['Total_salary']>ur,ur,data1['Total_salary'])
```

```python
data1['Total_salary']=np.where(data1['Total_salary']<lr,lr,data1['Total_salary'])
```
